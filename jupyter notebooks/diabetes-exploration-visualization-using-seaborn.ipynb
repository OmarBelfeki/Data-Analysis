```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/diabetes-dataset/diabetes_v3.csv")
data.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(data=data, x="Glucose", hue="Outcome", kde=True, element="step")
plt.title('Glucose Distribution by Outcome');
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 2)
sns.histplot(data=data, x="BMI", hue="Outcome", kde=True, element="step")
plt.title('BMI Distribution by Outcome');
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 3)
sns.boxplot(data=data, x="Outcome", y="Age")
plt.title('Age Distribution by Outcome');
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 4)
sns.boxplot(data=data, x="Outcome", y="Insulin")
plt.title('Insulin Distribution by Outcome');
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(data=data, x="Glucose", hue="Outcome", kde=True, element="step")
plt.title('Glucose Distribution by Outcome')

plt.subplot(2, 2, 2)
sns.histplot(data=data, x="BMI", hue="Outcome", kde=True, element="step")
plt.title('BMI Distribution by Outcome')

plt.subplot(2, 2, 3)
sns.boxplot(data=data, x="Outcome", y="Age")
plt.title('Age Distribution by Outcome')

plt.subplot(2, 2, 4)
sns.boxplot(data=data, x="Outcome", y="Insulin")
plt.title('Insulin Distribution by Outcome')

plt.tight_layout()
plt.show()
```
